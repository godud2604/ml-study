1. pandas를 이용해 csv 파일 읽어오기
2. 데이터프레임 정보 확인
    - 탐색적 자료 분석(EDA)를 통해 얻은 인사이트를 바탕으로 전처리한 변수, 연, 월, 일이 담겨 있는 데이터 셋 불러오기.
3. 요일 정보 추가
    - 요일 변수 추가
    - pd.to_datetime() 함수에 변환하고자 하는 날짜 데이터를 전달하여 날짜 형식으로 변환 후, .dt.day_name() 속성을 사용하여 날짜의 요일을 추출
4. 요일별 따릉이 대여량 비교
    - 요일별로 따릉이 사용량은 큰 차이가 없는 것으로 보임.
5. 요일별 + 연도별 따릉이 대여량 비교
    - hue 매개변수를 이용하여 연도(year)별로 그래프의 색상을 구분
    - 주중, 주말에 따른 따릉이 대여량은 큰 변화가 없었으며, 연도별로 구분했을 때도 큰 차이를 관찰하지 못했다.
6. 더위 점수(hot_score : high_temp * humidity)
    - 더위 점수(날씨가 덥고 습하다)가 따릉이 대여량에 영향을 주는 것으로 판단되며, 따릉이 대여량과 더위점수는 양의 상관관계임을 알 수 있다.
7. 추위 점수(cold_score : wind_speed / low_temp)
    - 추위 점수와 따릉이 대여량 사이엔 음의 상관관계가 있다.
    - 추위 점수(=날씨가 춥고 풍속이 강하다)가 따릉이 대여량에 영향을 주는 것으로 판단되므로, 추위 점수 변수를 생성해야 함.
8. 연도별 대여량
    - 연도별로 따릉이 대여량이 꾸준히 증가하는 것을 확인할 수 있다.
    - 따릉이는 연중무휴 서비스이기 때문에, 중간중간에 있는 아주 낮은 점들은 비가 많이 오는 등 날씨가 궂은 날이었다고 판단할 수 있다.
9. feature와 target의 상관관계
    - method="person" 은 피어슨 상관 계수를 사용하여 상관 행렬을 계산한다는 의미.
    - 온도와 따릉이 대여량은 양의 상관관계를 가진다.
    - 강수량과 미세먼지는 따릉이 대여량과 음의 상관관계를 가진다.
    - 결론적으로, 따릉이 대여량은 온도가 높을수록 증가하며, 강수량과 미세먼지가 낮을수록 증가하는 경향
10. 더위 점수 파생변수 생성
    - 따릉이 대여량(target)과 상관 관계가 있는 파생변수인 더위 점수를 새로운 변수로 생성해준다.
11. 추위 점수 파생변수 생성
    - 따릉이 대여량(target)과 상관 관계가 있는 파생변수인 추위 점수를 새로운 변수로 생성해준다.
